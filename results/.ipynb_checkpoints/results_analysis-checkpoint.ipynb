```python
import pandas as pd
import os
import json
```

```python
singlehead_dirs = ["./wheelchair_logs/singlehead/gadf/20230719-004908", "./wheelchair_logs/multihead/gasf/20230719-004907",
                 "./wheelchair_logs/multihead/mtf/20230719-004913", "./wheelchair_logs/multihead/rp/20230713-182132"]

multihead_df = pd.DataFrame(columns = ["encoding", "subject", "accuracy", "macro_f1", "weighted_f1"])

encoding = []
subject = []
accuracy = []
macro_f1 = []
weighted_f1 = []

for i in multihead_dirs:
    for j in os.listdir(i):
        if os.path.isdir(f"{i}/{j}"):
            f = open(f"{i}/{j}/score_dict.json")
            obj = json.load(f)

            encoding.append(i.split("/")[2])
            subject.append(j)
            accuracy.append(obj["accuracy"])
            macro_f1.append(obj["macro avg"]["f1-score"])
            weighted_f1.append(obj["weighted avg"]["f1-score"])


pd.DataFrame.from_dict({
    "encoding": encoding,
    "subject": subject,
    "accuracy": accuracy,
    "macro_f1": macro_f1,
    "weighted_f1": weighted_f1
}).to_csv("./multihead.csv")
```
